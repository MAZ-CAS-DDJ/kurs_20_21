# Ein kleines praktisches Tool JQ


[jq](https://stedolan.github.io/jq/) ist ein kleines command line tool um mit JSON zu arbeiten. Ihr k√∂nnt euch jq einfach [hier runterladen](https://stedolan.github.io/jq/download/) und auf eurem Mac installieren. Es gibt auch  [jqplay](https://jqplay.org/) um jq commands auszuprobieren. 

JSON macht sich drauf und dran CSV als Datenformat zu ersetzen, vor allem wenn es sich um un-strukturierte Daten handelt. Tools wie [twarc](https://github.com/edsu/twarc) dumpen ihre resultate auch als JSON.  

```python
DATA="st√§ndemehr.json"
```

## Einfaches Filtern

Man kann sich √§hnlich wie in pandas die ersten zeilen mit  `head` anzeigen lassen oder auch mit`cat` die ganze datei ausgeben lassen. 

```python
!head -1 $DATA | jq '.'
```

Wenn man nur die __Werte__ jedes fields haben m√∂chte:

```python
!head -1 $DATA| jq '.[]'
```

Man kann auch die Daten auf ausgew√§hlte Felder einschr√§nken:

```python
!head -3 $DATA | jq '[.created_at, .full_text]'
```

Die Twitter API beschreibt wie die Resultate aus dem [streaming](https://dev.twitter.com/streaming/overview) und [REST](https://dev.twitter.com/rest/public) API aussehen. 

Wie so oft bei JSON sind die Daten hierarchisch, d.h. dicts of dicts, so hat z.B. das `user` Feld daten √ºber den Nutzer der getweetet hat:

```python
!head -1 $DATA | jq '[.user]'
```

Wir k√∂nnen uns auch nur einen Teil der user fields ausgeben lassen:

```python
!head -2 $DATA | jq '[.user.screen_name, .user.name, .user.followers_count, .user.id_str]'
```

Manche felder tauchen mehrmals auf wie z.B. Hashtags oder Mentions. Man kann sie sich rausholen und zummen in einen String joinen:

```python
!cat $DATA | jq '[([.entities.hashtags[].text] | join(","))]'
```

# CSV Ouput

Man kann auch JSON daten in CSV umwandeln lassen. Mit der -r option werden die Felder als Strings f√ºr CSV exportiert.

```python
!head -8 $DATA | jq -r '[.id_str, .user.screen_name, .full_text] | @csv'
```

Das ganze l√§sst sich √ºber eine sog. pipe in eine Datei schreiben

```python
!cat $DATA | jq -r '[.id_str, .user.screen_name, .full_text] | @csv' > tweets.csv
```

```python
!head tweets.csv
```

# JSON Output

Man kann auch JSON als output w√§hlen um z.B: die Daten etwas vorzufiltern:

```python
!cat $DATA | jq -c '{{id: .id_str, user_id: .user.id_str, screen_name: .user.screen_name, created_at: .created_at, text: .text, user_mentions: [.entities.user_mentions[]?.screen_name], hashtags: [.entities.hashtags[]?.text], urls: [.entities.urls[]?.expanded_url]}}' > newtweets.json
```

```python
!head newtweets.json
```

# Datum
Man kann auch das Datum parsen:


```python
!head -n5 $DATA | jq -c '[.created_at, .created_at | strptime("%A %B %d %T %z %Y") | todate]'
```
